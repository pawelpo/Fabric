```python
import requests
from notebookutils.mssparkutils.credentials import getToken

def create_onelake_shortcut(sourceWorkspaceId, sourceItemId, sourcePath, path, directory):

    api_url = f"https://api.fabric.microsoft.com/v1/workspaces/{workspaceId}/items/{itemId}/shortcuts"

    base_url = "https://api.powerbi.com/v1.0/myorg"
    token = getToken("https://analysis.windows.net/powerbi/api")
    headers = {"Authorization": f"Bearer {token}"}

    json_body = {
        "path": f"{path}", 
        "name": f"{directory}",
        "target": {
            "oneLake": {
                "workspaceId": f"{sourceWorkspaceId}",
                "itemId": f"{sourceItemId}",
                "path": f"{sourcePath}"
            }
        }
    }

    try:
        response = requests.post(api_url, json=json_body, headers=headers)
        if response.status_code == 201:
            print("POST request successful.")
            print("Response:")
            print(response.json())
        else:
            print(f"POST request failed with status code: {response.status_code}")
            print("Response:")
            print(response.text)
    except Exception as e:
        print("An error occurred while sending the POST request:", str(e))

    return

workspaceId = spark.conf.get("trident.workspace.id") # current workspace ID
itemId = spark.conf.get("trident.lakehouse.id") # current lakehouse ID
path = "Tables" # Tables or Files
directory = "your_shortcut_name" # shortcut directory name (table name if located in Tables, folder name if in Files)

sourceWorkspaceId = "00000000-0000-0000-0000-000000000000" # target workspace ID
sourceItemId = "00000000-0000-0000-0000-000000000000" # target data item ID (e.g., a lakehouse)
sourcePath = "/Tables/your_existing_table" # path to target directory (e.g. /Tables/your_table_name or /Files/your_folder_name)

create_onelake_shortcut(sourceWorkspaceId, sourceItemId, sourcePath, path, directory)
```
